```python
import pandas as pd
import seaborn as sns
```

# Primeira Questão

```python
dados = pd.read_csv('dados/paginas_erros_vh.csv',sep=',')
```

## Qual o número médio de erros por página?

```python
dados['Erros'].mean()
```

## Qual a mediana de erros por página?

```python
dados['Erros'].median()
```

## Qual o desvio padrão de erros por página?

```python
dados['Erros'].std()
```

## Faça uma representação gráfica para a distribuição do número de erros


```python
sns.distplot(dados['Erros'],kde_kws={'clip': (0, 5)},bins=[0,1,2,3,4])
```

# Segunda Questão


```python
dados_2 = pd.read_csv('dados/dados_vh.csv',sep=',')
```

## Use cinco intervalos e construa um histograma

```python
sns.distplot(dados_2['Tamanho'],kde_kws={'clip': (2, 100)},bins=[0,20,40,60,80,100])
```

## Determine uma medida de posição central e uma medida de dispersão. 

Uma medida de posição central é uma forma de resumir os dados por meio de um único valor, por exemplo, se a mediana de um conjunto de dados é 10, isso significa que metade dos dados estão acima ou abaixo de 10.

Já medida de dispersão é utilizada para expor o quão espalhados são as informações reunidas, ou quão "diferentes" são os valores, por exemplo, quanto mais próximo de 0 for o desvio padrão, mais homogeneos são os dados.

É possivel observar os seguintes fatos:

- Não é possivel identificar outliers
